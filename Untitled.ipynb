```python
class TripleV2:
    def __init__(self,s,p,o):
        self.subject = s
        self.predicate = p
        self.object = o
        self.cost = 1
        self.matchtedKeyword = None
    
    def __str__(self):
        #return self.subject + '  ---  ' + self.predicate +'  ---  ' + self.object + '(' + str(self.cost) + ' : ' + str(self.matchtedKeyword) +')'
        return self.subject + '  ---  ' + self.predicate +'  ---  ' + self.object

    def getPredicate(self):
        return self.predicate
    
    def setCost(self, c):
        self.cost = c
        
    def getCost(self):
        return self.cost
    
    def getObject(self):
        return self.object
    
    def setMatchedKeyword(self,k):
        self.matchtedKeyword = k
```

```python
def createQueryString(label,sparql):
    th = 10000
    if countIncomingEdges(label,sparql) < th:
    #query = 'select distinct ?s ?p ?j ?k where {{?s ?p <' + label +'>} UNION {<' + label +'> ?j ?k}}'
    #filter wiki
        query = 'select distinct ?s ?p ?j ?k where {{?s ?p <' + label +'>. filter(!regex(str(?p), "wikiPage") && !regex(str(?p), "rdf-syntax"))} UNION {<' + label +'> ?j ?k. filter(!regex(str(?j), "wikiPage") && !regex(str(?j), "rdf-syntax"))}}'
    else:
        query = 'select distinct ?j ?k where {<' + label +'> ?j ?k. filter(!regex(str(?j), "wikiPage") && !regex(str(?j), "rdf-syntax"))}'
    return query
```

```python
def countIncomingEdges(obj,sparql):
    queryString = "SELECT count(DISTINCT ?s) as ?c WHERE { ?s ?p <" + obj +">}"
    sparql.setQuery(queryString)
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()
    #print(queryString)
    for result in results["results"]["bindings"]:
        return int(result['c']['value'])
```

```python
def fetchOneHop(entity, sparql):
    result_set = set()
    
    queryString = createQueryString(entity,sparql)
    sparql.setQuery(queryString)
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()
    #print(queryString)
    for result in results["results"]["bindings"]:
        if 's' in result:
            ansTp = TripleV2(entity, result['p']['value'], result['s']['value'])
            result_set.add(ansTp)
        else:
            ansTp = TripleV2(entity, result['j']['value'], result['k']['value'])
            result_set.add(ansTp)
    
    return result_set
```

```python
def fetchTwoHop(entity,sparql):
    result_two_hop = set()
    
    result_one_hop = fetchOneHop(entity,sparql)
    
    for triple in result_one_hop:
        result_two_hop.add(triple)
    
    for triple in result_one_hop:
        
        #expand only if the object starts with 'http://'
        triple_object = triple.getObject()
        if triple_object.startswith('http://'):
            current_two_hop = fetchOneHop(triple_object, sparql)

            for twp_hop_triple in current_two_hop:
                result_two_hop.add(twp_hop_triple)
    
    return result_two_hop
```
